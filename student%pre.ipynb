```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **loading the dataset**

```python
df=pd.read_csv("/content/Student Dataset final.csv")
df.head()
```

# **preprocessing**

```python
df.isnull().sum().sum()
```

```python
df.mean().round(1)
```

```python
df1=df.fillna(df.mean())
df1
```

```python
df1.isnull().sum().sum()
```

```python
df1.describe()
```

```python
df1.info()
```

# spliting the dataset

```python
x=df1.iloc[:,2:7]
y=df1.iloc[:,-1]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=4)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
y_train
```

## **taking algorithm**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()


lr.fit(x_train, y_train)

y_pred=lr.predict(x_test)




print("test accuracy",lr.score(x_test,y_test))
print("train accuracy",lr.score(x_train,y_train))



```

```python
pd.DataFrame(np.c_[x_test, y_test, y_pred], columns = ["sem1","sem2","sem3","sem4","sem5","sem6","predict sem6"])
```
